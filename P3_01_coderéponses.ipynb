```python
# QUESTION 3

# to create a copy of dataframe
p = p3.population[['code_pays','population']].copy()
d = p3.dispo_alim[['code_pays','pays','produit','dispo_alim_kcal_p_j','dispo_prot']].copy()

# merge : many to left : 1 to right : external left
d = d.merge(p, how= 'left', left_on = 'code_pays' , right_on = 'code_pays')

# to compute 'dispo énergétique' 
d['dispo_alim_kcal_p_j'] *= d['population'] * 365 

# to compute 'dispo protéine'
d['dispo_prot'] *= d['population'] * 365 / 1000

# to rename columns
d.rename(columns={'dispo_alim_kcal_p_j':'dispo_kcal',
                  'dispo_prot'         :'dispo_prot_kg'},inplace=True)

# to order columns
d = d[['code_pays','pays','produit','dispo_kcal','dispo_prot_kg']]

# to create index with useless rows
index_d_0 = d[ (d['dispo_kcal']==0)
              &(d['dispo_prot_kg']==0)
             ].index

# to drop useless rows
d.drop(index=index_d_0, inplace=True)

print("QUESTION 3 : ")
d
```

```python
#QUESTION 12

# SELECT FROM
exportations_15 = p3.balance_com[['produit','exportations','code_pays']].copy()

# WHERE 
exportations_15 = exportations_15[exportations_15['code_pays'].isin(p3.sous_nutrition['code_pays'])]

# to drop restriction's column : code_pays
del exportations_15['code_pays']

# GROUP BY : SOMME sur exportations REGROUPE par produit
exportations_15 = p3.group_by_reset_index(exportations_15,'produit','sum')

# ORDER BY exportations DESC  LIMIT 15
exportations_15 = p3.order_by_desc_limit(exportations_15,'exportations',15)


# SELECT FROM
importations_200 = p3.balance_com[['produit','importations']].copy()

# WHERE code_produit IN liste_produit_15_exportations
importations_200 = importations_200[importations_200['produit'].isin(exportations_15['produit'])]

# GROUP BY : produit WITH sum VALUES importations
importations_200 = p3.group_by_reset_index(importations_200,['produit'],'sum')

# ORDER BY importations DESC LIMIT 200
importations_200 = p3.order_by_desc_limit(importations_200,'importations',200)


# to compute ratios 
for index, product_name in enumerate(importations_200['produit']):
    
    importations_200.at[index,'ratio1'] = p3.donner_proportion ('autres_utilisations',
                                                                'dispo_int',
                                                                False,
                                                                p3.code_produit(product_name)
                                                               )

    importations_200.at[index,'ratio2'] = p3.donner_proportion ('alim_ani',
                                                                'nourriture',
                                                                True,
                                                                p3.code_produit(product_name)
                                                               )

ratio1 = p3.order_by_desc_limit(importations_200[['produit','ratio1']],'ratio1',3)
ratio2 = p3.order_by_desc_limit(importations_200[['produit','ratio2']],'ratio2',3)

print('QUESTION 12 :')
print("ratio 1 : proportion autres utilisations  /  disponibilité intérieure")
print("ratio 2 : proportion alimentation animale / (alimentation animale + humaine)")

ratio1.merge(ratio2, left_index=True, right_index=True)
```

```python
# QUESTION 13 

# SELECT|FROM
e = p3.equilibre_prod[['alim_ani','code_produit','pays']].copy()

# WHERE : pays USA  AND code_produit IN liste_cereales
e = e[ (e['pays'] == "États-Unis d'Amérique") & (e['code_produit'].isin(p3.Liste.cereales()))]

# SELECT|FROM
b = p3.balance_com[['productions','pays','code_produit']].copy()

# WHERE : pays USA  AND code_produit IN produits_animaux
b = b[ (b['pays'] == "États-Unis d'Amérique") & (b['code_produit'].isin(p3.Liste.produits_animaux()))]

ratio =  e['alim_ani'].sum() / b['productions'].sum() 

# to give freed cereals for 10% of productions
cereal_quantity = ( e['alim_ani'].sum() * 0.1 ) * ratio
# from MT to T
cereal_quantity *= 1000

print("QUESTION 13 : quantité de céréales libérées : {:.0f} Tonnes".format(cereal_quantity))
```
